```python
from PIL import Image
import os
from pathlib import Path

# กำหนดโฟลเดอร์ต้นทาง และโฟลเดอร์ปลายทาง
input_folder = '/home/nuttacha/archivenew/custom/pluss'
output_folder = 'custom'

# สร้างโฟลเดอร์ปลายทางหากยังไม่มี
#os.makedirs(output_folder, exist_ok=True)

# วนลูปผ่านไฟล์ทั้งหมดในโฟลเดอร์
for filename in os.listdir(input_folder):
    if filename.endswith(('.jpg', '.jpeg', '.png', '.WEBP')):
        # เปิดภาพ
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path)
        
        # ปรับไซส์เป็น 200x200
        custom = img.resize((200, 200))
        
        # บันทึกลงโฟลเดอร์ปลายทาง
        output_path = os.path.join(output_folder, filename)
        custom.save(output_path)
        
        print(f'ปรับไซส์เสร็จสิ้น: {filename}')

print('ทำการปรับไซส์ภาพทั้งหมดเสร็จเรียบร้อยแล้ว')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
import os

data_dir = "/home/nuttacha/archivenew/asl_alphabet_train/asl_alphabet_train"

#โหลด dataset และแบ่ง train / validation
train_ds = image_dataset_from_directory( #โดยอ่านภาพจากโฟลเดอร์ โดย image_dataset_from_directory
    data_dir,                                #หาโฟลเดอร์ย่อยใน data_dir เป็น class
    validation_split=0.2,     # 80% สำหรับ train
    subset="training",
    seed=123, 
    image_size=(128, 128), #โหลดรูป ปรับขนาดเป็น image_size RedsNet50V2 มันจะเป็น 224×224 ต้องลดเพราะเครื่องไม่ไหว จะลดข้อมูลเชิงพื้นที่ (spatial) จับรายละเอียดลดลงเล็กน้อย 
    batch_size= 4 #จำนวนภาพในแต่ละ batch ถ้าเพิ่มก็การฝึกเสถียร ลด noise ใน gradient แต่ได้แค่นี้เพราะ RAM ไม่พอ
    #เพิ่มเติม BatchNorm/LayerNorm พฤติกรรมจะเปลี่ยนตาม batch size — batch size ที่เล็กมากอาจทำให้ BatchNorm ทำงานไม่ดี
)

val_ds = image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(128, 128),
    batch_size= 4
)

class_names = train_ds.class_names
print("Class names:", class_names)
print("Number of classes:", len(class_names))

# Normalize ค่า pixel [0,1] # Normalize ค่า pixel [0,1] โมเดลเรียนได้ดีขึ้นเมื่อข้อมูลถูก normalized (ค่ากระจายไม่กว้างเกินไป) — ช่วยให้การเรียนรู้ของ optimizer มีเสถียรภาพ
normalization_layer = tf.keras.layers.Rescaling(1./255)
train_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
val_ds = val_ds.map(lambda x, y: (normalization_layer(x), y))

# Optimize dataset pipeline เก็บ dataset ไว้ในหน่วยความจำ (หรือไฟล์ชั่วคราว) เพื่ออ่านเร็วขึ้นใน epoch ถัดไป
'''train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE) 
val_ds = val_ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE)'''

# Optimize dataset pipeline
train_ds = train_ds.shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE) #สับข้อมูล train ให้ลำดับเปลี่ยนwไม่ให้โมเดลจำลำดับเดิม
val_ds = val_ds.prefetch(buffer_size=tf.data.AUTOTUNE) #เตรียมข้อมูล batch ต่อไปไว้ก่อน # AUTOTUNEให้ TensorFlow ปรับค่าที่เหมาะอัตโนมัติ

#ดูจำนวน batch
print("Train batches:", tf.data.experimental.cardinality(train_ds).numpy())
print("Validation batches:", tf.data.experimental.cardinality(val_ds).numpy())
```

```python
from tensorflow.keras.applications import ResNet50V2 #คลาสสำหรับโหลดสถาปัตยกรรม ResNet50V2
from tensorflow.keras import layers, models, optimizers #helper ของ Keras สำหรับสร้างเลเยอร์ โมเดล และตัว optimizer
import matplotlib.pyplot as plt #สำหรับวาดกราฟผลลัพธ์

# สร้าง base model (ใช้ pretrained weights จาก ImageNet)
base_model = ResNet50V2(
    include_top=False,         # ไม่เอา layer สุดท้ายของ ImageNet
    weights='imagenet',        # ใช้ pretrained weights
    input_shape=(128, 128, 3) #ขนาดภาพที่โมเดลรับเข้ามา: กว้าง=128 สูง=128 ช่องสี=3 (RGB)
)

# Freeze base model (ไม่ให้ปรับ weight ในช่วงแรก)
'''เราต้องการให้โมเดลเรียนเฉพาะส่วนหัว (head) ก่อน เพื่อให้ classifier ใหม่คุ้นกับ feature ที่ base_model ให้มา 
โดยไม่ไปเปลี่ยน feature พื้นฐานที่ pretrained ให้ไว้ (ลดความเสี่ยง overfitting และประหยัดเวลา)'''
base_model.trainable = False

# สร้างโมเดลใหม่ (Custom Head) เราต้องการ head ที่รับฟีเจอร์จาก backbone แล้วแปลงเป็นการคาดการณ์คลาสของงานเรา
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(512, activation='relu'), #เพื่อช่วยให้ training เสถียร
    layers.Dropout(0.3), #ช่วยลด overfitting มากขึ้น แต่ถ้ามากเกินไปจะทำให้ underfit
    layers.Dense(len(class_names), activation='softmax')  # จำนวน class = จำนวนโฟลเดอร์
])

# Compile โมเดล
model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001), #เริ่มต้น LR 0.001 เป็นค่าที่เหมาะสำหรับฝึก layer บนสุด (head)
    loss='sparse_categorical_crossentropy', #loss สำหรับ classification เมื่อ labels เป็น integer (ไม่ต้อง one-hot encode)
    metrics=['accuracy'] #ติดตาม accuracy ระหว่างเทรน/val
)

# ดูโครงสร้างโมเดล
model.summary()

# เทรนโมเดล (phase 1: fine-tune เฉพาะชั้นบน) ฝึกโมเดล (fit) บน train_ds และประเมินบน val_ds เป็นเวลา 10 epochs
history = model.fit( #เก็บค่าสถิติ (loss, accuracy) ของแต่ละ epoch
    train_ds,
    validation_data=val_ds,
    epochs=10
)

# Unfreeze บางส่วนของ ResNet เพื่อลอง fine-tune ต่อ
base_model.trainable = True  #train ได้ทั้งโมเดลโดยรวม
fine_tune_at = 140  # ปลดล็อกจาก layer ที่ 140 ขึ้นไป เริ่มให้ train (unfreeze) จากชั้น 140 ขึ้นไป

for layer in base_model.layers[:fine_tune_at]: #ปกติชั้นด้านล่างเรียน features พื้นฐาน (edges, textures) ที่เป็นสากล จึงไม่อยากแก้ไขมาก ส่วนชั้นบนจะเป็นฟีเจอร์เฉพาะที่เหมาะกับงานของเรา จึง fine-tune เฉพาะชั้นบน
    layer.trainable = False 

# Compile ใหม่ด้วย learning rate ที่ต่ำลง
model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-5), #คอมไพล์โมเดลใหม่เมื่อเปลี่ยนสถานะ trainable ของพารามิเตอร์ และเปลี่ยน learning rate เป็น 1e-5
    loss='sparse_categorical_crossentropy', #R ต่ำ: เมื่อ fine-tune layer ที่ pretrained มาแล้ว ต้องปรับค่า weights อย่างระมัดระวัง (ช้า ๆ) เพื่อไม่ให้ทำลายความรู้เดิม
    metrics=['accuracy']
)

# เทรนต่อ (phase 2)
'''ฝึกต่อจนถึง total_epochs โดยตั้ง initial_epoch=10 เพื่อให้การบันทึกประวัติ (history) ต่อเนื่อง

เหตุผล: ทำให้ history ที่ได้จาก phase1 และ phase2 รวมกันต่อเนื่อง (เพื่อ plot epoch ต่อเนื่อง)'''
fine_tune_epochs = 5
total_epochs = 10 + fine_tune_epochs

history_fine = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=total_epochs,
    initial_epoch=10
)

# รวมผลการเทรนสองช่วง เอาค่า accuracy ของ phase1 และ phase2 มาต่อกันเป็นลิสต์เดียว เพื่อ plot ต่อเนื่อง
acc = history.history['accuracy'] + history_fine.history['accuracy']
val_acc = history.history['val_accuracy'] + history_fine.history['val_accuracy']

# แสดงกราฟ Accuracy
plt.figure(figsize=(8, 4))
plt.plot(acc, label='Train Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title("ResNet50V2 Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# Save โมเดลไว้ใช้ภายหลัง ไว้ไป predic
model.save("resnet50v2_asl_model.h5")
print("Model saved as resnet50v2_asl_model.h5")

#
```
